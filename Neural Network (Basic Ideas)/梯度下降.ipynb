#### 单个变量的梯度下降

把梯度下降想成，在斜坡上随便丢一个球，这个球受到重力影响，会往低处滚，最后会滚到一个局部最低点。而梯度下降就是用类似方法帮助函数找到一个局部最小点，但它没办法找到全局最小点

- 随机选一个起始点θ(0)，当做初始值；
- 然后在θ(0)的地方，把cost function对θ(0)做微分，即dC(θ(0)) / dθ，做微分就是算切线的斜率。如果切线斜率为负，球会往右(下)滚，而θ(0)的值则会增加。反过来，如果切线斜率为正，则球会往左(上)滚，θ(0)的值会减小。因为切线斜率为负的，所以使用减号，这样θ(0)的值增加，球往下滚。η为学习速率，表示从θ(0)走到θ(1)的距离，即步长

$$ \theta ^{1}=\theta ^{0} - \eta dC(\theta ^{0}) \ /\  d\theta  $$

- 重复之前步骤，在计算θ(1)的斜率，斜率为负，于是继续减去，往下滚，直到到达局部最小值

#### 两个变量的梯度下降

- 假设θ有两个值（θ(1)，θ(2)），同样随机选一个起始向量，当做初始向量；

$$ \theta ^{0} = \begin{bmatrix}
\theta _{1}^{0}\\ 
\theta _{2}^{0}
\end{bmatrix} $$

- 计算损失函数C(θ)在θ(0)位置的梯度（如果θ(0)只有一个值，则是斜率），即计算cost function对θ(1)做偏微分，这个就是cost function在θ(1)这个地方的梯度；对θ(2)做偏微分，是cost function在θ(2)这个地方的梯度。

$$ \nabla C(\theta ^{0}) = \begin{bmatrix}
\partial C(\theta _{1}^{0})\  /\  \partial \theta_{1} \\ 
\partial C(\theta _{2}^{0})\  /\  \partial \theta_{2}
\end{bmatrix} $$

- 然后更新参数，因为θ有两个值（θ(1)，θ(2)），所以梯度也有两个值，是二维的向量，更新参数方式如下，让球继续往低处滚：

$$
\begin{bmatrix}
\theta _{1}^{1}\\ 
\theta _{2}^{1}
\end{bmatrix} 
=
\begin{bmatrix}
\theta _{1}^{0}\\ 
\theta _{2}^{0}
\end{bmatrix}
-
\eta 
\begin{bmatrix}
\partial C(\theta _{1}^{0})\  /\  \partial \theta_{1}\\ 
\partial C(\theta _{2}^{0})\  /\  \partial \theta_{2}
\end{bmatrix}
\Rightarrow 
\theta ^{1} = \theta ^{0} - \eta \nabla C(\theta^{0})
$$

- 接着跟之前一样，计算θ(1)的梯度，更新参数。一直重复下去，直到达到局部最小值

$$ \nabla C(\theta ^{1}) = \begin{bmatrix}
\partial C(\theta _{1}^{1})\  /\  \partial \theta_{1} \\ 
\partial C(\theta _{2}^{1})\  /\  \partial \theta_{2}
\end{bmatrix} $$
