#### 结构化学习，输入和输出都为结构化对象。
y不像回归和分类是数字，而是一些复杂结构，如系列、边界框等。举例：
- 语音识别领域，X: 语音信号(序列) -> Y: 文本(序列)
- 翻译领悟，X: 汉语句子(序列) -> Y: 英语句子(序列)
- 语法分析领域，X: 序列 -> Y: 解析树(树结构)
- 对象检测领域，X: 图片 -> Y: 边界框
- 概况领域，X: 长的文档 -> Y: 总结概况(短的段落)
- 检索领域，X: 关键词 -> Y: 搜索结果(网页列表)

#### 结构化学习步骤：
- training，寻找一个函数F，输入为X和Y，输出为实值R，即F(X, Y) = R。F(X, Y)评估X和Y是否相容，举例X为一篇长文档，Y为一段总结概括。如果R很高，代表Y这段概括能很好表达出这篇长文档X。同样，X为汉语句子，Y为英语句子。如果F(X, Y)的值大， 表明X能被恰当地翻译，即X和Y相容，则F(X, Y)的值高；
- testing，给定一个x，找出所有y中与F(x, y)相容性最高的y，类似于P(y | x)，给定x寻找y值，使得概率最大化。
